```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt
from math import sqrt

from sklearn.metrics import mean_absolute_error
from sklearn.metrics import mean_squared_error

%matplotlib inline
import seaborn as sns ; 


df = pd.read_csv('conso_train.csv', sep=';',  usecols=['date', 'puissance'] )
df.columns = ['date','puissance']


df.head()
print(df)

#Pour éviter les glitchs sur la date de météo
df.date = df.date.str.replace('h', ':')
df.date = pd.to_datetime(df.date, format='%Y-%m-%dT%H:%M:%S')
df.date = df.date.dt.strftime('%Y-%m-%d %H:%M')

df.date = pd.to_datetime(df.date)
df.index = df.date

#Creating puissance set 
puissance=df[0:] 


df.Timestamp = pd.to_datetime(df.date,infer_datetime_format=True)
df.index = df.Timestamp 

puissance.Timestamp = pd.to_datetime(puissance.date,infer_datetime_format=True) 
puissance.index = puissance.Timestamp 
```

```python
puissance.puissance.plot(figsize=(15,8), title= 'Consommation observée', fontsize=14)
plt.show()
```

```python
y_hat_avg = puissance.copy()
y_hat_avg['avg_forecast'] = puissance['puissance'].mean()
plt.figure(figsize=(12,8))
plt.plot(puissance['puissance'], label='Consommation observée')
plt.plot(y_hat_avg['avg_forecast'], label='Average Forecast')
plt.legend(loc='best')
plt.show()

error=mean_absolute_error(puissance.puissance, y_hat_avg.avg_forecast)
print('MAPE:%.3f'% error)

rmse = sqrt(mean_squared_error(puissance.puissance, y_hat_avg.avg_forecast))
print('RMSE: %.3f' % rmse)
```

```python
#Changement des données prises en compte (ici on prend le mois d'avril à titre d'exemple)

avril = df[4799:5519] 
```

```python
def predictConsoPerDay():
    predTab =dict()    
    res = []
    i=0
    for i in range(0,len(df)):
        day = df.date[i].weekday()
        predTab.setdefault(day,[]).append(df.puissance[i])
        
    for k,v in predTab.items():
        predTab[k]=np.average(v)
    
    for key in range(0,len(predTab)):
        res.append(predTab[key])    

    print(res)
            
```

```python
predConsoPerDay = predictConsoPerDay()
```

```python
def predictConsoPerHourPerDay():
    predTab =dict()    
    i=0
    res = []
    for i in range(0,len(df)):
        hour = df.date[i].hour
        predTab.setdefault(hour,[]).append(df.puissance[i])
        
    for k,v in predTab.items():
        predTab[k]=np.average(v)
        
    for key in range(0,len(predTab)):
        res.append(predTab[key])    

    #print(res)
    return (res)
            
```

```python
predConsoPerHPerD = predictConsoPerHourPerDay()
```

```python
sns.tsplot(data=list(predConsoPerHPerD))
```

```python
predConsoPerHPerDForAMonth = 30 * predictConsoPerHourPerDay()
```

```python
sns.tsplot(data=list(predConsoPerHPerDForAMonth), color='r')

sns.tsplot(data=list(avril.puissance))

error=mean_absolute_error(avril.puissance, predConsoPerHPerDForAMonth)
print('MAPE:%.3f'% error)

rmse = sqrt(mean_squared_error(avril.puissance, predConsoPerHPerDForAMonth))
print('RMSE: %.3f' % rmse)
```

```python
def predictConsoPerHourPerWeek():
    predTab =dict() 
    i=0
    res = []
    for i in range(0,len(df)):
        hour = (df.date[i].weekday()*24) + df.date[i].hour
        predTab.setdefault(hour%168,[]).append(df.puissance[i])   
        
            
    for k,v in predTab.items():
        predTab[k]=np.average(v)
        
    for key in range(0,len(predTab)):
        res.append(predTab[key])    

   ## print(res)
    return (res)
            
```

```python
avrilWeek = df[4799:5471]
```

```python
predConsoPerHourPerWeekForAMonth=predictConsoPerHourPerWeek()*4
```

```python
sns.tsplot(data=list(predConsoPerHourPerWeekForAMonth), color='r')

sns.tsplot(data=list(avril.puissance))

error=mean_absolute_error(avrilWeek.puissance, predConsoPerHourPerWeekForAMonth)
print('MAPE:%.3f'% error)

rmse = sqrt(mean_squared_error(avrilWeek.puissance, predConsoPerHourPerWeekForAMonth))
print('RMSE: %.3f' % rmse)
```

```python
def predictConsoPerHourPerDayForAGivenMonth(m):
    predTab =dict()    
    i=0
    res = []
    for i in range(0,len(df)):
        if(m == df.date[i].month):
            hour = df.date[i].hour
            predTab.setdefault(hour,[]).append(df.puissance[i])
        
    for k,v in predTab.items():
        predTab[k]=np.average(v)
        
    for key in range(0,len(predTab)):
        res.append(predTab[key])    

    #print(res)
    return (res)
            
```

```python
predConsoPerHourPerDayForApril=predictConsoPerHourPerDayForAGivenMonth(4)*30
```

```python
sns.tsplot(data=list(predConsoPerHourPerDayForApril), color = 'r')

sns.tsplot(data=list(avril.puissance))

error=mean_absolute_error(avril.puissance, predConsoPerHourPerDayForApril)
print('MAPE:%.3f'% error)

rmse = sqrt(mean_squared_error(avril.puissance, predConsoPerHourPerDayForApril))
print('RMSE: %.3f' % rmse)
```
