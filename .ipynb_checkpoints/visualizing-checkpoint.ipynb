```python
from datetime import datetime
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.dates as mdates

conso = pd.read_csv('conso_train.csv', sep=';')
meteo = pd.read_csv('meteo_train.csv', sep=';', usecols=['Date UTC', 'T¬∞ (C)', 'P (hPa)', 'HR (%)', 'P.ros√©e (¬∞C)', 'Visi (km)', 'Vt. moy. (km/h)','Vt. raf. (km/h)', 'Vt. dir (¬∞)','RR 3h (mm)','Neige (cm)','Nebul. (octats)'])

display(conso.head())
display(meteo.head())
```

```python
meteo.columns = ['date','temp','pression','hr','pros','visi','vmoy','vraf','vdir','rr','neige','nebul']
```

```python
conso.date = pd.to_datetime(conso.date)
conso.index = conso.date

meteo.date = meteo.date.str.replace('h', ':')
meteo.date = pd.to_datetime(meteo.date, format='%d/%m/%y %H:%M')
meteo.date = meteo.date.dt.strftime('%Y-%m-%d %H:%M')
meteo.date = pd.to_datetime(meteo.date)
meteo.index = meteo.date

display(conso.head())
display(meteo.head())
```

```python
meteo.temp.plot(figsize=(19,9), title='Température', fontsize=14, label='Train', secondary_y=True)


#Resample de la puissance suivant la date par jour (1 point au lieu de 24)
#conso = conso.resample('D').mean()
conso.puissance.plot(figsize=(19,9), title='Consommation', fontsize=14, label='Train')

plt.show()


meteo.pression.plot(figsize=(19,9), title='Pression', fontsize=14, label='Train')
plt.show()
meteo.hr.plot(figsize=(19,9), title='Humidité relative', fontsize=14, label='Train')
plt.show()
meteo.pros.plot(figsize=(19,9), title='Pros', fontsize=14, label='Train')
plt.show()
meteo.visi.plot(figsize=(19,9), title='Visibilité', fontsize=14, label='Train')
plt.show()
meteo.vmoy.plot(figsize=(19,9), title='Vitesse moyenne vent', fontsize=14, label='Train')
plt.show()
meteo.vraf.plot(figsize=(19,9), title='Vitesse rafale vent', fontsize=14, label='Train')
plt.show()
meteo.vdir.plot(figsize=(19,9), title='Vitesse ?', fontsize=14, label='Train')
plt.show()
```
